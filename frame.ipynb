```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from collections import OrderedDict
```

```python
def show_frame(img):
    """ Render frame inline with matplotlib """
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

```python
# Create VideoCapture object from input file
cap = cv2.VideoCapture('videos/2_RUE_540x960.mp4')  
cap = cv2.VideoCapture('videos/39_MJ_540x960.mp4') 
vd = OrderedDict()
i = 0

while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
        vd[i] = frame
    else: 
        break
    i += + 1
print(len(vd), 'frames loaded')

# ------------------------------------------------ #
#            Append first frame to end
# ------------------------------------------------ #
print('Append first frame to end')
vd[i] = vd[0]
print(len(vd), 'frames')
print('\nVerify:', (vd[i] == vd[0]).all())
```

```python
# ------------------------------------------------ #
#                 Write out to MP4
# ------------------------------------------------ #

# Default resolutions of the frame are obtained.The default resolutions are system dependent.
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
print('\nOutput', frame_width, 'x', frame_height)

# Define the codec and create VideoWriter object.The output is stored in 'outpy.avi' file.
out = cv2.VideoWriter('videos/output.mp4',
                      cv2.VideoWriter_fourcc('M','J','P','G'), 24, 
                      (frame_width,frame_height))

# Write the frame into the file 'output.avi'
for k in range(i+1):
    out.write(vd[k])
    
print('\t', len(vd), 'frames')
# release video and write objects
out.release()
cap.release()
```
